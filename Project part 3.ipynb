## I - Introduction
	This dataset was taken from the paper Functional traits data for testate amoebae of Northern Holarctic realm. This paper isn’t testing any sort of hypothesis, but was instead published to provide a comprehensive dataset on the physical characteristics of all known testate (shelled) amoeba species located in North America, Europe, and Russia. The data was compiled from other existing published datasets.


## II - Research questions, data, and figures
- Each observation in this case is one species of testate amoeba; 372 total. Each is an average of multiple observations of that species taken from other data sets; those originals directly measured individual amoebas taken from soil samples.
- There are 30 features, being genus, minimum shell length, , average shell length, minimum shell width, maximum shell width, average shell width, minimum shell depth, maximum shell depth, average shell depth, R1, R2, shell (test) shape, minimum aperture length, maximum aperture length, average aperture length, minimum aperture width, maximum aperture width, average aperture width (in µm) , R3, R4, position of the aperture, degree of invagination of aperture, aperture rim, presence/absence of collar, shell covering, presence/absence of internal partitions, presence/absence of spines or horns, and feeding type. These all tell us the averages of traits for the species.
- The dataset was created entirely from other ones that took direct data on amoeba morphology, both from other researchers as well as the authors themselves.
- The datasets made from on samples taken in Russia (which comprised the majority of samples) came from the authors, another dataset was used from samples taken in Europe, and another on samples taken in North America.
-  Descriptions of the features themselves were given for each paper, so standardizing them between them was not difficult. However, descriptions of the methods behind measuring each feature were not given.
- This article doesn’t have a hypothesis. Instead it aims to be a useful tool to other researchers by compiling other data on testate amoebae, which the authors say are morphologically diverse, ecologically important, and good indicators of ecosystem health. No other comprehensive dataset like this existed before this.
- One could ask “How does an amoeba species’s feeding type correlate with average shell width?” as the data set provides a comprehensive overview of shelled amoeba, allowing a qualitative variable to be compared with a degree of significance.

## III - Data cleaning and exploration

#### 3.1: Load and clean data
The dataset was taken directly from the published paper; the .xlsx file and its explanatory pdf was is available under reference 37 (or available at figshare.com/articles/dataset/species-level_trait_csv/26955565/2). The excel file was exported as a .csv file and saved to the same directory that other .ipynb files are accessed from. 

```python
# This line imports the pandas module into jupyter, which is needed for dataframe operations (such as below)
import pandas as pd
import seaborn as sns
#This line loads our .csv data file from the paper, converts it to a dataset using pandas, and assigns it to a variable so that it can be altered later.
dfspecies = pd.read_csv("species-level_trait(species-level_trait).csv", index_col = 0)
dfspecies.head(3)
```

Some steps had to be taken to clean the data (remove features with null values)

```python
#First a dataframe was created that contains all rows (species) that contained a null value for miniumum shell length
nullspecies = pd.isnull(dfspecies['Minimum shell length'])
#This dataframe was subtracted from the base dataframe and saved to a variable, removing null values for that column
aspecies = dfspecies.loc[~nullspecies, :]
#This process was repeated for all features that contained null values
nullspecies = pd.isnull(aspecies['Maximum shell length'])
bspecies = aspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(bspecies['Average shell length'])
cspecies = bspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(cspecies['Minimum shell width'])
dspecies = cspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(dspecies['Maximum shell width'])
especies = dspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(especies['Average shell width'])
fspecies = especies.loc[~nullspecies, :]
nullspecies = pd.isnull(fspecies['Minimum shell depth'])
gspecies = fspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(gspecies['Maximum shell depth'])
hspecies = gspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(hspecies['Average shell depth'])
ispecies = hspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(ispecies['R1'])
jspecies = ispecies.loc[~nullspecies, :]
nullspecies = pd.isnull(jspecies['R2'])
kspecies = jspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(kspecies['Shell (test) shape'])
lspecies = kspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(lspecies['Minimum aperture length'])
mspecies = lspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(mspecies['Maximum aperture length'])
nspecies = mspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(nspecies['Average aperture length'])
ospecies = nspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(ospecies['Minimum aperture width'])
pspecies = ospecies.loc[~nullspecies, :]
nullspecies = pd.isnull(pspecies['Maximum aperture width'])
qspecies = pspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(qspecies['Average aperture width'])
rspecies = qspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(rspecies['R3'])
sspecies = rspecies.loc[~nullspecies, :]
nullspecies = pd.isnull(sspecies['R4'])
cleanspecies = sspecies.loc[~nullspecies, :]
```

#### 3.2: Describe data numerically
The code below gives the shape of the cleaned dataframe, or the number of rows by the number of columns. In this case the number of rows is equal to the number of observations (species) and the number of columns is equal to the number of species.

```python
print(cleanspecies.shape)
```

The "describe" function outputs a series of 8 (or 4 if the feature is categorical) different statistical datapoints on a given feature. For each given feature, the output for each statistic would be the following:
- ___Average shell depth___
- - __Predicted:__
- - count = 335, exactly. This is just a count of all observations.
  - mean = around 70. 3 observations have this value. Definitionally, this is the average.
  - std = around 40. Observations appear to have a mostly even distribution between roughly 20 and 100, making deviation large.
  - min = 5. This appears to be the lowest value.
  - 25% = around 30. This is the first quartile. By eyeballing it, 1 in 4 species appear to have a depth below this value
  - 50% = around 60. Aka the median. This will be lower than the mean, as there are a few very large outliers. There are 3 observations with this value.
  - 75% = around 80. 3rd quartile; one in 3 datapoints appear to be higher.
  - max = 272. This was the largest value there seems to be.
  - __Actual:__

```python
cleanspecies['Average shell depth'].describe()
```

- - The count was exactly correct, but nothing else was. There was at least one higher value for the min and lower value for the max. Q1 and Q3 were mostly accurate, but the mean, median, and std were about 10 points off.

- ___Genus___
- - __Predicted:__
- - count = 335, exactly. Again this is a count of all observations.
  - unique = about 80. This is the number of qualitative variables present, roughly eyeballed from the previous figure regarding species diversity.
  - top = "Difflugia". top is the most frequent value, which the paper directly states to be difflugia for genuses.
  - freq = about 70. This is the frequency of the top value, eyeballed here to be 70.
  - __Actual:__

```python
cleanspecies['Genus'].describe()
```

- - The number of unique genuses was overestimated, but everything else is about right.

### 3.3: Visualizations.
__First visualization:__ A univariate histogram of average shell depth per species. A histogram provides a view of the distribution of a feature; as such it would be useful to use one to get an idea of the average size of each species. 

```python
#creates a displot (histogram) from the cleaned species dataset, with the averae shell depth set as the x-axis.
sns.displot(data=cleanspecies, x="Average shell depth")
```

- This graph makes the mean and median given from the describe function from earlier more obvious, and easier to see. The minimum and maxiumum are also clearly visible.
- The paper appears to have a graph displaying the same data but with different bins.

__Second visualization:__ A multivariate (3 variables in this case) scatterplot comparing apeture length to apeture width, with the genus of each species as the hue. A scatterplot will show the correlation between two variables if one exists (which one likely does for the width and length of the same thing). This will also provide more data on each genus, information that is lacking in the paper itself.

```python
#Creates a relplot (scatter plot) with the x axis set to the apeture length, the y axis the apeture width, and the hue as the genus (assigned default colors)
sns.relplot(data=cleanspecies, x="Average aperture length", y="Average aperture width", hue = "Genus")
```

- The wide variety of colors (and the huge legend) is expected from the value of 57 unique species from the describe function output.
- This data shows the expected pattern of apeture width correlating with apeture length, however it also shows a different pattern: many ameoba species in the dataset obstensively have the exact same values for both features. Given that every species bucking that trend shows a greater length than width, this most likely points to many of the datasets this information was compiled from being disengenuous (or at least not thorough enough) with their observations.
- The inability to completely trust all measurements of each species is indeed a problem the paper mentions, as the methods of each dataset these were taken from are not completely known.
